Question 3: Data Reduction/Bucketing. First Variable:PriceUSD

```python
airline.PriceUSD.describe()
```

```python
buckets = [ 30.990, 1055.660, 2393.865, 5395.000, 7000.000, 9924.120 ]
bucketlabels = ['Cheap', 'Not too Bad', 'Medium Expensive', 'Expensive', 'Outrageous']
airline['buckets'] = pd.cut(airline['PriceUSD'] , bins=buckets, labels=bucketlabels, include_lowest=True)
airline.head()
```

```python
airline.groupby(['buckets']).count()
```

```python
plt.hist(airline['buckets'], bins=3)
plt.show()
```

Second Variable: Flight Distance

```python
airline.FlightDistance.describe()
```

```python
buckets = [ 31, 835, 1723, 2860, 3300, 4983 ]
bucketlabels = ['Instant', 'Short', 'Medium', 'Upper Medium', 'Long']
airline['buckets'] = pd.cut(airline['FlightDistance'] , bins=buckets, labels=bucketlabels, include_lowest=True)
airline.head()
```

```python
airline.groupby(['buckets']).count()
```

```python
plt.hist(airline['buckets'], bins=3)
plt.show()
```

Explanation: When looking at data no matter if it is big or small, it is just data until you are able to draw conclusions and interpret it. With bucketing it allows you to break apart a larger collection of data into smaller bits in which you can find potential patterns or useable information. It would be interesting to see this bucketing method towards U.S gas prices over the past 60 years in relation to the involvement of the United States in the middle east at the time. This component allows you to take large amounts of numbers and narrow the focus to topics that can be used for beneficial applications.
